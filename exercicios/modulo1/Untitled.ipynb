```python
import pandas as pd 
pd.set_option('display.max_columns', 21)
arquivo = pd.read_csv(r'/Users/gustavohenriquebonatopiresdemoraes/ia-portfolio-gustavo/exercicios/modulo1/kc_house_data.csv')
arquivo.head(15)
```

```python
delete_columns = ['id', 'date', 'zipcode', 'lat', 'long']
arquivo.drop(columns = delete_columns, inplace= True)
arquivo.shape
```

```python
y = arquivo['price']
x = arquivo.drop('price', axis = 1)
```

```python
# A regressão linear é uma técnica estatística e de machine learning que modela a relação entre uma variável dependente 
#(a que queremos prever) e uma ou mais variáveis independentes (as que usamos para prever) como uma equação linear

from sklearn.linear_model import LinearRegression
modelo = LinearRegression()
```

```python
from sklearn.model_selection import train_test_split
x_teste, x_treino, y_teste, y_treino = train_test_split(x, y, test_size = 0.30, random_state = 14) 
```

```python
modelo.fit(x_treino, y_treino)
resultado = modelo.score(x_teste, y_teste)
print(resultado)
```

```python
# Regularização L2 (Ridge) A Regressão Ridge é uma técnica estatística usada para corrigir o overfitting (sobreajuste) em modelos de machine learning
#e regressão linear, especialmente em casos de multicolinearidade (alta correlação entre variáveis)

from sklearn.linear_model import Ridge
# é padrão o alpha = 1
modeloRidge = Ridge(alpha=1)
modeloRidge.fit(x_treino, y_treino)
resultadoRidge = modeloRidge.score(x_teste, y_teste)
print(resultadoRidge)
```

```python
#Regularização L1 (Lasso) zera coeficientes – seleciona variáveis.
from sklearn.linear_model import Lasso
modeloLasso = Lasso(alpha = 30, max_iter = 50000, tol= 0.1)
modeloLasso.fit(x_treino, y_treino)
resultadoLasso = modeloLasso.score(x_teste, y_teste)
print(resultadoLasso)
```

```python
#Elastic Net Regression é um algoritmo de machine learning supervisionado que combina as técnicas de regularização Lasso (L1) e Ridge (L2) 
#para realizar regressão
from sklearn.linear_model import ElasticNet
modeloElastic = ElasticNet(alpha = 0.5, l1_ratio = 0.5, tol = 0.3)
modeloElastic.fit(x_treino, y_treino)
resultadoElastic = modeloElastic.score(x_teste, y_teste)
print(resultadoElastic)
```
